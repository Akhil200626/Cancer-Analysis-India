```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("cancer_india.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x="Cancer_Type", data=df)
plt.title("Distribution of Cancer Types in India")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x="Survival_Status", data=df)
plt.title("Survival Status of Cancer Patients")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x="Cancer_Type", hue="Tobacco_Use", data=df)
plt.title("Cancer Type vs Tobacco Use")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x="Stage", data=df, order=["I","II","III","IV"])
plt.title("Cancer Stage Distribution")
plt.show()
```
